```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
df = pd.read_csv('kc_house_data.csv')
df
```

### 1.2. Изучите стоимости недвижимости

```python
df['price'].min(), df['price'].max()
```

```python
plt.figure(figsize = (10,4))

plt.hist(df['price'], bins = 20)
plt.title('Стоимость недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.xticks(range(75000, 8000000, 200000), rotation=45)
plt.grid();
```

### Выводы:
- Максимальное количествоо недвижимости продается в диапазоне цен от 75 000 до 450 000.
- Основные продажи ограничиваются ценой примерно 850 000.
- Дома с ценой выше 2 000 000 продаются крайне редко.

### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize = (6,4))

plt.hist(df['sqft_living'], bins = 15)
plt.title('Распреление жилой квадратуры')
plt.xlabel('Жилые квадратные метры')
plt.ylabel('Количество')
plt.xticks(range(100, 14000, 500), rotation=45)
plt.grid();
```

### Выводы:
- Максимальное количествоо недвижимости продается с жилой площадью от 1 000 до 2 000 квадратных метров.
- Основные продажи ограничиваются домами с жилой площадью примерно 3 600 квадратных метров.
- Дома с жилой площадью выше 6 000 продаются крайне редко.

### 1.4. Изучите распределение года постройки

```python
data_yr = df['yr_built'].value_counts(ascending=True)
data_yr.head()
```

```python
years_list = [1900+i*10 for i in range (0,13)]
```

```python
names = data_yr.index
values = data_yr.values
yr_list = list(range(1900,2025,5))
plt.figure(figsize=(16, 6))
plt.bar(names,values, width=0.7, facecolor='g', alpha=0.75)
plt.title('Годы постройки домов')
plt.xlabel('Годы')
plt.ylabel('Количество домов')
plt.grid(True)
plt.xticks(yr_list,rotation=40);
plt.show()
   
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
sns.lineplot(data=data_yr, x=names,y=values)
plt.xlabel('Годы')
plt.ylabel('Количество домов')
plt.xticks(ticks=years_list);
```

### Выводы:
- Кризисы строительтва были в середине 1930-х и начале 1970-х
- Пики строительства были в конце 1960-х, в конце 1970-х, и середине нулевых

### 2.2. Изучите распределение этажей домов

```python
df.floors.max(), df.floors.min()
```

```python
data2 = df['floors'].value_counts(normalize=True)*100
data2
```

```python
data_floor = df['floors'].value_counts()
f_val = data_floor.values
f_ind = data_floor.index
plt.figure(figsize=(12, 6))
plt.bar(f_ind,f_val, width=0.4, facecolor='c', alpha=0.75)
plt.title('Этажность')
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,8))
explode = [0.0, 0.0, 0.0, 0.0, 0.3, 0.4]
plt.pie(data2.values, autopct='%1.2f%%', labels=data2.index, explode=explode)
plt.legend(data2.index)
plt.title('Распределение этажей домов');
```

### Выводы:
- Большая часть домов имеет один или два этажа;
- Одноэтажные дома лидируют, их доля составляет 49,41%.
- Минимальная доля продаж приходится на 2,5-этажные и 3,5-этажные дома (0,74%, 0,04%)

### 2.3. Изучите распределение состояния домов

```python
data_cond = df['condition'].value_counts(sort=False)
data_cond
cond_val= data_cond.values
cond_ind = data_cond.index
cond_explode = [0,0,0,0.6,0.3]
colors = ["#ff7f0e", "#9467bd", "#d62728","#1f77b4", "#8c564b"]
plt.title("Состояние домов")
plt.pie(cond_val, labels=cond_ind, radius = 1.5,
        startangle=-20, colors=colors,
        rotatelabels=True, explode = cond_explode,
        labeldistance=1.1, autopct='%.1f%%')
plt.legend(data_cond.index)
plt.title("Состояние домов", bbox={'facecolor':'0.8', 'pad':5})
plt.show;
```

### Выводы:
- Домов в плохом состоянии (1 и 2) - менее 1%
- Домов в среднем состоянии (3) - 64.9%
- Домов в хорошем состоянии (4) - 26.3%
- Домов в отличном состоянии (5) - 7.9%
